```python
import pandas as pd
import numpy as np
import datetime as dt
from sklearn.decomposition import PCA
```

```python
data =  pd.read_csv("swissyield.csv")
```

```python
date = pd.to_datetime(data['Date'], format="%Y/%m/%d")
data.set_index(date, inplace = True)
data.drop(['Date'], axis = 1, inplace = True)
data1 = data.diff()
```

```python
data1.head()
```

```python
inputdf = data1[data1.index > dt.datetime(2000,8,31)]
inputdf.head()
```

```python
pca = PCA(n_components=5)
pca.fit(inputdf)
print(pca.explained_variance_ratio_)

pca2 = PCA(n_components = 3)
pca2.fit(inputdf)
print(pca2.explained_variance_ratio_)
print(pca.singular_values_)
print(pca.components_)
```

```python
#df = inputdf.transpose()
```

```python
muhat = np.array(inputdf.mean(axis =0))
muhat
```

```python
covhat = inputdf.cov()
covhat
```

```python
from numpy.linalg import eig
values, vectors = eig(covhat)
values
```

```python
vectors
```

```python
print(values[0])
print(values[1])

print(vectors[0])
print(vectors[1])

vectors = pd.DataFrame(vectors)
vectors
```

```python
PCs = np.asmatrix(inputdf - muhat) * np.asmatrix(vectors)
PCdf = pd.DataFrame(PCs)
PCdf.set_index(inputdf.index, inplace = True)
PCdf.rename(columns = dict(zip(PCdf.columns, inputdf.columns)), inplace = True)
PCdf.cov()
```
